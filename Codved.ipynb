```python
# TASK 1 (LEVEL 1) : DATA CLEANING AND DATA PREPROCESSING

import pandas as pd

# Load the dataset using pandas (iris dataset)
df = pd.read_csv('Downloads/1) iris.csv')
print("\nFirst 5 rows are:\n", df.head())

# Missing values & duplicates
print("\n Missing values are:\n", df.isnull().sum())
df = df.drop_duplicates()
```

```python
#TASK 2 (LEVEL 1) : EDA

# Summary statistics 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("\n Mean values:\n", df.mean(numeric_only=True))
print("\n Median values:\n", df.median(numeric_only=True))
print("\n Mode values:\n", df.mode(numeric_only=True))
print("\n Standard Deviation:\n", df.std(numeric_only=True))

# Visualizations

# Histogram of Sepal Width
plt.figure(figsize=(10,5))
sns.histplot(df['sepal_width'], kde=True, color='darkblue')
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()

# Boxplot of petal_width
plt.figure(figsize=(10,5))
sns.boxplot(x=df['petal_width'], color='purple')
plt.title("Boxplot of Petal Width")
plt.xlabel("Petal Width")
plt.show()

# Scatter plot between sepal_length and petal_length
plt.figure(figsize=(10,5))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
